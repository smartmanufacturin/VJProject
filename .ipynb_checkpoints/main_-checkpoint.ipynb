```python
import os
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm  # Progress bar

# ===== Configuration =====
DATASET_PATH = "./cityscapes/"
OUTPUT_PATH = "cityscapes_masks"
MODE = "train"  # "train" or "val"
MASK_TYPE = "multiclass"  # "binary" or "multiclass"

# Cityscapes class definitions (19 classes + background)
CLASSES = {
     7: "road", 8: "sidewalk", 11: "building", 12: "wall", 13: "fence",
    17: "pole", 19: "traffic light", 20: "traffic sign", 21: "vegetation",
    22: "terrain", 23: "sky", 24: "person", 25: "rider", 26: "car",
    27: "truck", 28: "bus", 31: "train", 32: "motorcycle", 33: "bicycle"
}

# ===== Create Output Directories =====
os.makedirs(f"{OUTPUT_PATH}/{MODE}", exist_ok=True)

# ===== Process Each Image =====
image_dir = f"{DATASET_PATH}/leftImg8bit/{MODE}"
label_dir = f"{DATASET_PATH}/gtFine/{MODE}"

cities = os.listdir(image_dir)

for city in tqdm(cities, desc="Processing cities"):
    city_img_dir = f"{image_dir}/{city}"
    city_label_dir = f"{label_dir}/{city}"
    
    for img_name in os.listdir(city_img_dir):
        if "_leftImg8bit.png" not in img_name:
            continue
            
        # Load image and corresponding label
        img_path = f"{city_img_dir}/{img_name}"
        label_path = f"{city_label_dir}/{img_name.replace('_leftImg8bit.png', '_gtFine_labelIds.png')}"
        
        if not os.path.exists(label_path):
            print(f"Missing label for {img_name}")
            continue
            
        # Read label image (contains class IDs per pixel)
        label = np.array(Image.open(label_path))
        
        # Initialize mask
        if MASK_TYPE == "binary":
            mask = np.zeros_like(label, dtype=np.uint8)
            mask[label > 0] = 1  # 1=object, 0=background
        else:  # multiclass
            mask = label  # Already contains class IDs
        
        # Save mask
        mask_name = img_name.replace("_leftImg8bit.png", "_mask.png")
        cv2.imwrite(f"{OUTPUT_PATH}/{MODE}/{mask_name}", mask)

print("Mask generation complete!")
```

```python
import matplotlib.pyplot as plt

def visualize_cityscapes(img_path, mask_path):
    img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title("Input Image")
    
    plt.subplot(1,2,2)
    plt.imshow(mask, cmap='nipy_spectral', vmin=0, vmax=19)
    plt.title("Segmentation Mask")
    plt.colorbar()
    plt.show()

# Example usage
img_sample = "cityscapes/leftImg8bit/train/aachen/aachen_000005_000019_leftImg8bit.png"
mask_sample = "cityscapes_masks/train/aachen_000005_000019_mask.png"
visualize_cityscapes(img_sample, mask_sample)
```

```python
import os
import json
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm
from matplotlib import pyplot as plt

# ========================
# Configuration
# ========================
DATASET_PATH = "./cityscapes/"
OUTPUT_PATH = "processed_cityscapes"
MODE = "train"  # or "val"

# Cityscapes 19-class labels (official)
CLASSES = {
    7: "road",
    8: "sidewalk", 
    11: "building",
    12: "wall",
    13: "fence",
    17: "pole",
    19: "traffic light",
    20: "traffic sign",
    21: "vegetation",
    22: "terrain",
    23: "sky",
    24: "person",
    25: "rider",
    26: "car",
    27: "truck",
    28: "bus",
    31: "train",
    32: "motorcycle",
    33: "bicycle"
}

# ========================
# Create Output Directories
# ========================
os.makedirs(f"{OUTPUT_PATH}/images/{MODE}", exist_ok=True)
os.makedirs(f"{OUTPUT_PATH}/masks/{MODE}", exist_ok=True)
os.makedirs(f"{OUTPUT_PATH}/annotations", exist_ok=True)

# ========================
# Initialize COCO-style Annotation
# ========================
coco_ann = {
    "info": {"description": "Cityscapes processed dataset"},
    "licenses": [],
    "categories": [{"id": k, "name": v} for k, v in CLASSES.items()],
    "images": [],
    "annotations": []
}

image_id = 1
annotation_id = 1

# ========================
# Process Each Image
# ========================
image_dir = f"{DATASET_PATH}/leftImg8bit/{MODE}"
label_dir = f"{DATASET_PATH}/gtFine/{MODE}"

for city in tqdm(os.listdir(image_dir), desc=f"Processing {MODE} cities"):
    city_img_dir = f"{image_dir}/{city}"
    city_label_dir = f"{label_dir}/{city}"
    
    for img_name in os.listdir(city_img_dir):
        if not img_name.endswith("_leftImg8bit.png"):
            continue
            
        # ========================
        # 1. Process Image
        # ========================
        img_path = f"{city_img_dir}/{img_name}"
        base_name = img_name.replace("_leftImg8bit.png", "")
        
        # Copy image to output
        output_img_path = f"{OUTPUT_PATH}/images/{MODE}/{base_name}.jpg"
        img = cv2.imread(img_path)
        cv2.imwrite(output_img_path, img)
        
        # Add to COCO annotations
        height, width = img.shape[:2]
        coco_ann["images"].append({
            "id": image_id,
            "file_name": f"{base_name}.jpg",
            "width": width,
            "height": height
        })
        
        # ========================
        # 2. Process Mask and Annotations
        # ========================
        label_path = f"{city_label_dir}/{base_name}_gtFine_labelIds.png"
        if not os.path.exists(label_path):
            continue
            
        label = np.array(Image.open(label_path))
        
        # Create mask (only include valid classes)
        mask = np.zeros_like(label, dtype=np.uint8)
        for class_id in CLASSES.keys():
            mask[label == class_id] = class_id
            
        # Save mask
        cv2.imwrite(f"{OUTPUT_PATH}/masks/{MODE}/{base_name}_mask.png", mask)
        
        # ========================
        # 3. Create Instance Annotations
        # ========================
        for class_id in CLASSES.keys():
            class_mask = (label == class_id).astype(np.uint8)
            
            if np.sum(class_mask) == 0:  # Skip if class not present
                continue
                
            # Find contours (for polygon annotation)
            contours, _ = cv2.findContours(
                class_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            
            for contour in contours:
                if contour.shape[0] < 3:  # Skip invalid polygons
                    continue
                    
                # Simplify polygon
                epsilon = 0.001 * cv2.arcLength(contour, True)
                approx = cv2.approxPolyDP(contour, epsilon, True)
                
                # Flatten coordinates [x1,y1,x2,y2,...]
                segmentation = approx.flatten().tolist()
                
                # Add to COCO annotations
                coco_ann["annotations"].append({
                    "id": annotation_id,
                    "image_id": image_id,
                    "category_id": class_id,
                    "segmentation": [segmentation],
                    "area": float(cv2.contourArea(contour)),
                    "bbox": list(cv2.boundingRect(contour)),  # [x,y,width,height]
                    "iscrowd": 0
                })
                annotation_id += 1
                
        image_id += 1

# ========================
# Save Annotations
# ========================
with open(f"{OUTPUT_PATH}/annotations/instances_{MODE}.json", "w") as f:
    json.dump(coco_ann, f)

print(f"\nProcessing complete! Saved to {OUTPUT_PATH}")

# ========================
# Visualization
# ========================
def visualize_sample(image_path, mask_path, annotation_path):
    """Visualize image + mask + annotations"""
    img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    with open(annotation_path) as f:
        ann = json.load(f)
    
    # Create annotation overlay
    overlay = img.copy()
    for obj in ann["annotations"]:
        contour = np.array(obj["segmentation"][0]).reshape(-1, 2)
        cv2.drawContours(overlay, [contour], -1, (255, 0, 0), 2)
    
    plt.figure(figsize=(18, 6))
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title("Input Image")
    
    plt.subplot(1, 3, 2)
    plt.imshow(mask, cmap="jet")
    plt.title("Segmentation Mask")
    
    plt.subplot(1, 3, 3)
    plt.imshow(overlay)
    plt.title("Annotations Overlay")
    plt.show()

# Example visualization
sample_base = list(coco_ann["images"])[0]["file_name"].replace(".jpg", "")
visualize_sample(
    f"{OUTPUT_PATH}/images/{MODE}/{sample_base}.jpg",
    f"{OUTPUT_PATH}/masks/{MODE}/{sample_base}_mask.png",
    f"{OUTPUT_PATH}/annotations/instances_{MODE}.json"
)
```
